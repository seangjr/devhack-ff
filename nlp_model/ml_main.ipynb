```python
import re
#manipulation of regular expressions
```

```python
def cleaning_resume_dataset(your_dataset):
    your_dataset = re.sub('RT|cc', ' ', your_dataset) #remove RT and CC
    your_dataset = re.sub('#\S+', ' ', your_dataset)
    your_dataset = re.sub('@\S+', ' ', your_dataset)
    your_dataset = re.sub('http\S+\s*', ' ', your_dataset) #remove http urls
    your_dataset = re.sub('https\S+\s*', ' ', your_dataset) #remove https urls
    your_dataset = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', your_dataset) #remove punctuations
    your_dataset = re.sub(r'[^\x00-\x7f]',r' ', your_dataset) #non ASCII characters
    your_dataset = re.sub('\s+', ' ', your_dataset)
    return your_dataset

resume_dataset['Cleaned_resume'] = resume_dataset.Resume.apply(lambda x: cleaning_resume_dataset(x))
resume_dataset
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('stopwords')

#these are all for the tokenization process, which is essentially the process of breaking down larger sentences into smaller chunks
```

```python
def preprocess_text(text):
    text = text.lower()
    text = ''.join([char for char in text if char.isalpha() or char.isspace()])
    words = word_tokenize(text)
    words = [word for word in words if word not in set(stopwords.words('english'))]
    lemmatized_words = [token.lemma_ for token in nlp(' '.join(words))]
    text = ' '.join(lemmatized_words)
    return text

resume_dataset['Cleaned_resume'] = resume_dataset.Cleaned_resume.apply(lambda x: preprocess_text(x))
jobs_dataset['Cleaned_jdesc'] = jobs_dataset.Cleaned_jdesc.apply(lambda x: preprocess_text(x))
```

```python
resume_embeddings = [nlp(text).vector for text in resume_dataset['Cleaned_resume']]
jobs_embeddings = [nlp(text).vector for text in jobs_dataset['Cleaned_jdesc']]

X_train, X_test, y_train, y_test = train_test_split(resume_embeddings, range(len(jobs_embeddings)), test_size=0.3, random_state=42)
# Split the data into training and testing sets (70/30 ratio)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(X_train, jobs_embeddings)
```

```python
best_matches = []
for i in range(len(similarity_scores)):
    best_match_index = similarity_scores[i].argmax()
    best_match_score = similarity_scores[i][best_match_index]
    best_matches.append((i, best_match_index, best_match_score))

#Finding the best match for the resume in the training set 
```

```python
results_dataframe = pd.DataFrame(best_matches, columns=['Resume_index', 'Jobs_index', 'Match_score'])
#Creating a dataframe to store the results
```

```python
print("Training Set Results: ")
results_dataframe
```

```python
test_similarity_scores = cosine_similarity(X_test, jobs_embeddings)
```

```python
test_best_matches = []
for i in range(len(test_similarity_scores)):
    best_match_index = test_similarity_scores[i].argmax()
    best_match_score = test_similarity_scores[i][best_match_index]
    test_best_matches.append((i, best_match_index, best_match_score))
```

```python
import fitz #you have to pip install PyMuPDF for this, fitz is a PyMuPDF library
```

```python
def pdf_to_text(pdf_path):
    try:
        with fitz.open(pdf_path) as pdf_document:
            text =""
            for page_number in range(pdf_document.page_count):
                page= pdf_document[page_number]
                text += page.get_text()
        return text
    except Exception as e:
        print(f"Error: {e}")
        return None
```

```python
input_resume_embeddings =  nlp(resume_extracted).vector
```

```python
cosine_similarities = cosine_similarity([input_resume_embeddings], jobs_embeddings)
```

```python
top_3_indices = cosine_similarities.argsort()[0][::-1][:3]
```

```python
top_3_matches = []
for index in top_3_indices:
    job_id = jobs_dataset.at[index, 'Job_ID']
    company_name = jobs_dataset.at[index, 'Company_name']
    match_score = cosine_similarities[0][index]
    job_title = jobs_dataset.at[index, 'Job_title']
    top_3_matches.append([job_id, company_name, job_title, match_score])

# Print or return the top 3 matches
top_3_matches_df = pd.DataFrame(top_3_matches, columns=['Job_ID', 'Company_name', 'Job_title', 'Match_score'])
print("Top 3 Job Matches:")
print(top_3_matches_df)
```
