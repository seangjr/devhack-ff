```python
#creating a list made up of the resume and job
full_text = [resume, job_desc] #creating matrix
```

```python
#counting the number of times a word appears blah blah blah
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
count_matrix =cv.fit_transform(full_text)
```

```python
#cosine_similarity from sklearn will calculate how similar the texts are 
from sklearn.metrics.pairwise import cosine_similarity
```

```python
print("These are your similarity scores AYO AYO: \n")
print(cosine_similarity(count_matrix))
```

```python
#[0][1] is the position that you wanna get from the matrix printed above
#[[ [0,0], [0,1] ]
#[ [1,0], [1,1]]]
convertPercentage = cosine_similarity(count_matrix)[0][1] * 100
convertPercentage = (round(convertPercentage, 0)) #0 for 0 decimal places
```
